## **Linear Quantization: A Popular Method**  

In **linear quantization**, we map **high-precision values** to **lower precision**, such as:  
- `FP32 â†’ INT8`  

### **Formula**  
The relationship between real values (`r`) and quantized values (`q`) is given by:  

```python
r = s * (q - z)
```

Where:  
- `s` â†’ Scale factor (from the original tensor)  
- `z` â†’ Zero point (from the quantized tensor)  

To compute `q` from `r`, we rearrange the formula:  

```python
q = (r / s) + z
```

Since `q` must be an integer, we apply rounding:  

```python
q = int(round(r / s + z))
```

### **Python Implementation**

```


````python
def linear_Q(
        tensor,
        scale,
        zero,
        dtype = torch.int8
):
    
    scaled_shifted_tensor = tensor / scale + zero 
    rounded = torch.round(scaled_shifted_tensor)

    q_min = torch.iinfo(dtype).min 
    q_max = torch.iinfo(dtype).max 

    q_tensor = rounded.clamp(q_min, q_max).to(dtype)

    return q_tensor


````

## **Quantization Error Visualization Helper Functions**

In this section, we define several helper functions to assist with visualizing the effects of quantization on tensors. These functions allow us to:

1. Calculate the **quantization error** between the original and dequantized tensors.
2. **Plot heatmaps** of the original, quantized, dequantized tensors, and the quantization error tensor.

These tools help in understanding how quantization impacts the precision of the tensor values, providing insight into the trade-offs between model size reduction and potential loss of information during quantization.


````python
# ploting helper function for quantization error 

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap

def quantization_error(tensor, dequantized_tensor):
    return (dequantized_tensor - tensor).abs().square().mean()

def plot_matrix(tensor, ax, title, vmin=0, vmax=1, cmap=None):
    """
    Plot a heatmap of tensors using seaborn
    """
    sns.heatmap(tensor.cpu().numpy(), ax=ax, vmin=vmin, vmax=vmax, cmap=cmap, annot=True, fmt=".2f", cbar=False)
    ax.set_title(title)
    ax.set_yticklabels([])
    ax.set_xticklabels([])


def plot_quantization_errors(original_tensor, quantized_tensor, dequantized_tensor, dtype = torch.int8, n_bits = 8):
    """
    A method that plots 4 matrices, the original tensor, the quantized tensor
    the de-quantized tensor and the error tensor.
    """
    # Get a figure of 4 plots
    fig, axes = plt.subplots(1, 4, figsize=(15, 4))

    # Plot the first matrix
    plot_matrix(original_tensor, axes[0], 'Original Tensor', cmap=ListedColormap(['white']))

    # Get the quantization range and plot the quantized tensor
    q_min, q_max = torch.iinfo(dtype).min, torch.iinfo(dtype).max
    plot_matrix(quantized_tensor, axes[1], f'{n_bits}-bit Linear Quantized Tensor', vmin=q_min, vmax=q_max, cmap='coolwarm')

    # Plot the de-quantized tensors
    plot_matrix(dequantized_tensor, axes[2], 'Dequantized Tensor', cmap='coolwarm')

    # Get the quantization errors
    q_error_tensor = abs(original_tensor - dequantized_tensor)
    plot_matrix(q_error_tensor, axes[3], 'Quantization Error Tensor', cmap=ListedColormap(['white']))

    fig.tight_layout()
    plt.show()
````

## **Scale and Zero Point Calculation for Quantization**

In linear quantization, we map the range of real numbers \([r_{\text{min}}, r_{\text{max}}]\) to a fixed range of integers \([q_{\text{min}}, q_{\text{max}}]\). The relationship is defined by the following formulas:

- For \(r_{\text{min}}\), \(r_{\text{max}}\):
  - `r_min = s(q_min - z)`
  - `r_max = s(q_max - z)`

Where:
- `s` is the **scale** factor, which ensures the range of real values fits into the quantized range.
- `z` is the **zero point**, which shifts the quantized values to align with the real value range.

We can derive the scale and zero point as follows:

- **Scale (s)**:  
  `s = (r_max - r_min) / (q_max - q_min)`
  
- **Zero Point (z)**:  
  `z = int(round(q_min - r_min / s))`

In case the calculated zero point `z` is out of the quantized range, we handle the edge cases by ensuring that `z` lies between `q_min` and `q_max`:
- If `z > q_max`, then set `z = q_max`.
- If `z < q_min`, then set `z = q_min`.

In the following code, we calculate the scale and zero point (`s`, `z`) based on a given real tensor, ensuring that we handle edge cases for `z` as well.


````python
def get_q_scale_and_zero_point(r_tensor, dtype=torch.int8):
    """
    Get quantization parameters (scale, zero point) 
    for a floating point tensor
    """
    q_min, q_max = torch.iinfo(dtype).min, torch.iinfo(dtype).max
    r_min, r_max = r_tensor.min().item(), r_tensor.max().item()

    scale = (r_max-r_min)/(q_max-q_min)

    zero_point = q_min-(r_min/scale)

    # clip the zero_point to fall in [quantized_min, quantized_max]
    if zero_point < q_min or zero_point > q_max:
        zero_point = q_min
    else:
        # round and cast to int
        zero_point = int(round(zero_point))
    return scale, zero_point
````

````python
def linear_q_with_scale_and_zero_point(
    tensor, scale, zero_point, dtype = torch.int8):

    scaled_and_shifted_tensor = tensor / scale + zero_point

    rounded_tensor = torch.round(scaled_and_shifted_tensor)

    q_min = torch.iinfo(dtype).min
    q_max = torch.iinfo(dtype).max

    q_tensor = rounded_tensor.clamp(q_min,q_max).to(dtype)
    
    return q_tensor
````

## **Asymmetric vs Symmetric Quantization**

When it comes to quantization, two main types are widely used: **Asymmetric Quantization** and **Symmetric Quantization**. Both methods reduce the precision of the model weights and activations, but they differ in how the real-value range is mapped to the quantized range.

<p align="center">
  <img src="images/symm.png" alt="Quantization Overview">
</p>

### **Asymmetric Quantization**

In **Asymmetric Quantization**, the range of the real values and the quantized values do not necessarily have the same zero point. This means that the zero point (i.e., the point representing zero in the real value range) might not align with zero in the quantized value range. The formula for asymmetric quantization is:

- **Scale (s)**:  
  `s = (r_max - r_min) / (q_max - q_min)`
  
- **Zero Point (z)**:  
  `z = int(round(q_min - r_min / s))`

Where:
- `r_min` and `r_max` are the minimum and maximum values of the real tensor.
- `q_min` and `q_max` are the minimum and maximum values in the quantized tensor.
- `s` is the scale factor.
- `z` is the zero point.

Asymmetric quantization is often used when the range of values in the original tensor doesn't symmetrically span around zero, and therefore, aligning the zero point between the real and quantized values may not be suitable.

### **Symmetric Quantization**

In **Symmetric Quantization**, the zero point of the real tensor aligns directly with the zero point of the quantized tensor. In this approach, both the real values and the quantized values are symmetric around zero. The formula for symmetric quantization is:

- **Scale (s)**:  
  `s = (r_max - r_min) / (q_max - q_min)`
  
- **Zero Point (z)**:  
  `z = 0`

In symmetric quantization, the real values are typically mapped to the quantized range symmetrically. This approach is commonly used when the real tensor values are symmetric around zero, such as in certain activation functions where negative values are as frequent as positive values.

### **Key Differences**

- **Asymmetric Quantization**:
  - Zero point in the real tensor does not necessarily match zero in the quantized tensor.
  - More flexible, as it can handle any distribution of the values in the real tensor.
  
- **Symmetric Quantization**:
  - Zero point in both the real tensor and the quantized tensor is aligned.
  - Simpler and works best when the real tensor values are symmetric around zero.

### **When to Use Which?**
- Use **Asymmetric Quantization** when the data does not have a symmetric distribution or when the zero point does not naturally align between real and quantized values.
- Use **Symmetric Quantization** when the data is symmetric, which is often the case with activations from neural networks where negative and positive values are distributed similarly.

Both quantization methods help to reduce model size and improve inference speed by lowering the bit-width of the model weights and activations.


````python
##SEMMETRICS VS ASYMMERTICS MODE : 

def get_q_scale_symmetric(tensor, dtype=torch.int8):
    r_max = tensor.abs().max().item()
    q_max = torch.iinfo(dtype).max

    # return the scale
    return r_max/q_max
````

## **Asymmetric vs Symmetric Quantization: Key Differences**

### **Asymmetric Quantization**
- **Full utilization of the quantized range**: When using asymmetric quantization, the entire quantized range is utilized, meaning both positive and negative values are mapped effectively, even if the real tensor does not have a symmetric distribution around zero.
  
### **Symmetric Quantization**
- **Bias handling**: In symmetric quantization, if the floating-point values are biased toward one side (e.g., in ReLU activations), the quantized range will also be biased toward that side. This could lead to suboptimal use of the quantized range if the values are heavily skewed.
  
- **Simplicity**: Symmetric quantization is generally simpler to implement since it aligns the zero point in both the real tensor and the quantized tensor, resulting in fewer computations.
  
- **Memory efficiency**: Symmetric quantization is more memory-efficient because there is no need to store the zero point (`z`), which is always set to 0. This saves memory space when dealing with large models.

### **Practical Usage**
- **For 8-bit quantization**: Symmetric quantization is commonly used because it is simple and efficient, and 8-bit precision is often enough for most deep learning models.
  
- **For lower bit-widths (2-bit, 4-bit)**: Asymmetric quantization is typically preferred when working with lower bit-widths, as it provides more flexibility to handle the reduced precision without losing too much information. The quantized range needs to be adjusted carefully to maintain performance at these lower bit depths.

---

## **Types of Quantization Methods**

1. **Per Tensor Quantization**  
   In per tensor quantization, the scale and zero-point are applied uniformly across the entire tensor. This approach is simpler but may not always capture fine-grained variations across different parts of the tensor.

2. **Per Channel Quantization**  
   Per channel quantization applies different scales and zero-points to each channel of the tensor (e.g., for each filter in a convolutional layer). This allows for more precision in cases where different channels have different distributions of values, leading to better performance.

---

### **Summary**
- **Symmetric Quantization**:
  - Simpler and more memory-efficient, as no zero-point storage is required.
  - Suitable for 8-bit quantization.
  - Works well when the data is symmetric or near symmetric.
  
- **Asymmetric Quantization**:
  - Utilizes the full quantized range.
  - Preferred for lower bit-widths (2 or 4-bit) where precision loss is critical.
  - Suitable when data is not symmetric or biased.

Both quantization methods offer trade-offs, and choosing the right method depends on the bit-width, memory constraints, and model characteristics.


## **Per Channel Quantization**

In **per channel quantization**, we perform quantization independently for each channel (or row/column, depending on your choice). This means we need to store a separate scale and zero-point value for each channel. As a result, **memory requirements are lower**, and this method is typically used with 8-bit quantization.

### Key Points:
- **Per Channel Quantization**: Scale and zero-point are computed for each channel individually (e.g., for each row or column, depending on your choice of dimension).
- **Memory Efficiency**: By storing separate scales and zero-points per channel, memory requirements are lower than for per tensor quantization.
- **Usage**: Commonly used for 8-bit quantization when preserving the precision of individual channels is necessary.

<p align="center">
  <img src="images/per_channel.png" alt="Quantization Overview">
</p>

Dont get intimidated by the function we will break down step by step:

````python
def linear_q_symmetric_per_channel(r_tensor, dim, dtype=torch.int8):
    
    output_dim = r_tensor.shape[dim]
    # store the scales
    scale = torch.zeros(output_dim)

    for index in range(output_dim):
        sub_tensor = r_tensor.select(dim, index)
        scale[index] = get_q_scale_symmetric(sub_tensor, dtype=dtype)

    # reshape the scale
    scale_shape = [1] * r_tensor.dim()
    scale_shape[dim] = -1
    scale = scale.view(scale_shape)
    quantized_tensor = linear_q_with_scale_and_zero_point(
        r_tensor, scale=scale, zero_point=0, dtype=dtype)
   
    return quantized_tensor, scale
````

## **Per Group Quantization**

In **per group quantization**, we apply quantization independently to different groups within a tensor. These groups could represent rows, columns, or any arbitrary division along a chosen dimension. Each group will have its own scale and zero-point value, making this approach flexible and more efficient in certain scenarios. This method allows the tensor to have varying precision across different regions or groups.

### Key Points:
- **Per Group Quantization**: Quantization is applied to smaller subgroups of a tensor (e.g., specific rows, columns, or custom groupings based on a chosen dimension).
- **Memory Efficiency**: Unlike per tensor quantization, this approach requires less memory for storing individual scales and zero-points for each group.
- **Flexibility**: It adapts well to tensors with varied value distributions, allowing for finer control over quantization at the group level.
- **Usage**: Often used when a tensor has regions with varying ranges or different types of data (e.g., different image regions with varied brightness or activations from different layers in a neural network).


<p align="center">
  <img src="images/per_group.png" alt="Quantization Overview">
</p>

On the side of practical implenmtaion we can utilize the linear_q_symmetric_per_channel function but we must make sure to the channel by groups number

````python
def linear_q_symmetric_per_group(tensor, group_size,
                                 dtype=torch.int8):
    
    t_shape = tensor.shape
    assert t_shape[1] % group_size == 0
    assert tensor.dim() == 2
    
    tensor = tensor.view(-1, group_size)
    
    quantized_tensor, scale = linear_q_symmetric_per_channel(
                                tensor, dim=0, dtype=dtype)
    
    quantized_tensor = quantized_tensor.view(t_shape)
    
    return quantized_tensor, scale
````

````python
def linear_dequantization_per_group(quantized_tensor, scale, 
                                    group_size):
    
    q_shape = quantized_tensor.shape
    quantized_tensor = quantized_tensor.view(-1, group_size)
    
    dequantized_tensor = linear_dequantization(quantized_tensor, 
                                               scale, 0)
    
    dequantized_tensor = dequantized_tensor.view(q_shape)
    
    return dequantized_tensor
````

Now lets make quantize and dequantize method , plus the workflow to quantize a model should as follow : 

we detect all the linear layer on the model 

we loop over thejm and replace the standard nn.linear with the custom made class 

````python
class W8A16LinearLayer(nn.Module):
    def __init__(self, in_features, out_features, bias=True, dtype=torch.float32):
        super().__init__()

        self.register_buffer("int8_weight",
                            torch.randint(-128, 127, (out_features, in_features), dtype=torch.int8))
        
        self.register_buffer("scale", torch.randn((out_features)
                             , dtype=dtype))
        
        if bias:
            self.register_buffer("bias", torch.zeros((1, out_features)
                                 , dtype=dtype))
        else:
            self.bias = None

    def quantize(self, weight):
        ## cast to fp32 is recommended for stability
        w_fp32 = weight.clone().to(torch.float32)

        scales = w_fp32.abs().max(dim=-1).values / 127 
        scales = scales.to(weight.dtype) 

        int8_weight = torch.round(w_fp32 / scales.unsqueeze(1)).to(torch.int8)

        self.int8_weight = int8_weight
        self.scale = scales

    def forward(self, input):
        return w8_16_forward(self.int8_weight, input, self.scale, self.bias)
    


````

Now we ready to make Quantizer (a quantization pipline) which will iterate over the original linear model and replace them with the int8  

````python
def replace_module(module,
                target_class, 
                module_name_to_exclude):
    
    for name, child in module.named_children():

        # if the layer is a linear layer and not in the exclusion list
        if isinstance(child, nn.Linear) and not any([x == name for x in module_name_to_exclude]):
            old_bias = child.bias
            new_module = target_class(
                child.in_features,
                child.out_features,
                child.bias is not None,
                child.weight.dtype
            )
            setattr(module, name, new_module)
            if old_bias is not None:
                getattr(module , name).bias = old_bias 
        
        else : 
            # recursively call the function
            replace_module(child, target_class, module_name_to_exclude)
````

````python
# LETS MAKE A DUMMY MODEL (LIKE MINI GPT)
class DummyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Embedding(16, 32)
        # with bias
        self.l2 = nn.Linear(32, 64)
        # without bias
        self.l3 = nn.Linear(64, 128, bias=False)
        self.l4 = nn.Linear(128, 256, bias=False)

````

````python
def replace_module_and_quantize(module,
                target_class, 
                module_name_to_exclude):
    
    for name, child in module.named_children():

        # if the layer is a linear layer and not in the exclusion list
        if isinstance(child, nn.Linear) and not any([x == name for x in module_name_to_exclude]):
            old_bias = child.bias
            old_weight = child.weight

            new_module = target_class(
                child.in_features,
                child.out_features,
                child.bias is not None,
                child.weight.dtype
            )
            setattr(module, name, new_module)
            getattr(module , name).quantize(old_weight)

            if old_bias is not None:
                getattr(module , name).bias = old_bias 
        
        else : 
            # recursively call the function
            replace_module_and_quantize(child, target_class, module_name_to_exclude)
````

````python
model_3 = DummyModel()

replace_module_and_quantize(model_3, W8A16LinearLayer, ["l4"])

print("quantized model:",model_3)
````

````python

def model_info(model):
    """
    Returns the number of parameters and size in MB of a PyTorch model.

    Args:
        model (torch.nn.Module): The PyTorch model.

    Returns:
        tuple: (num_params, size_in_MB)
    """
    num_params = sum(p.numel() for p in model.parameters())
    size_in_MB = sum(p.element_size() * p.numel() for p in model.parameters()) / (1024 * 1024)
    
    return num_params, size_in_MB


num_params, size_in_MB = model_info(model)
print(f"Number of parameters: {num_params}")
print(f"Size of model in MB: {size_in_MB}")
print(f"Model size: {model_size}")
print(f"size in BG: {size_in_MB/1024}")
````

## **What's Next? ðŸš€**

Having mastered linear quantization, it's time to take things further. The next step is exploring **State-of-the-Art (SOTA) methods** for scaling large models. These methods are designed to handle **outliers** and model weights that may not fit well within the constraints of simple linear quantization. For example:

- **Non-linear quantization** techniques, like **logarithmic quantization** and **adaptive quantization**, can better handle extreme outliers in the weight distribution.
  
- **Mixed-Precision Quantization**: In some layers, you may want to use higher bit precision (like 16-bit) to avoid significant performance degradation in critical sections of the model.

- **Dynamic Quantization** and **Quantization-Aware Training (QAT)**: These methods improve upon static quantization by incorporating quantization during the training process, ensuring that the model adapts to the reduced precision, leading to better accuracy retention.


## **Why Linear Quantization Fails for Certain Models**

Linear quantization works well for models with relatively uniform weight distributions, but in practice, many models, especially large ones, have **outliers**â€”weights that are significantly larger or smaller than the rest. Linear quantization does not handle these extreme values well, as it scales all the weights by the same factor. This can result in the loss of important information or precision.
